## Step 1: Import Libraries & Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better-looking plots
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

print("✓ Libraries imported successfully")
```

```python
# Load the dataset
df = pd.read_csv('telco_churn.csv')
print(f"Dataset loaded: {df.shape[0]} rows × {df.shape[1]} columns")
```

## Step 2: First Look at the Data

```python
# Display first few rows
print("\n=== FIRST 5 ROWS ===")
df.head()
```

```python
# Check data types
print("\n=== DATA TYPES ===")
print(df.dtypes)
```

```python
# Check for missing values
print("\n=== MISSING VALUES ===")
missing = df.isnull().sum()
if missing.sum() == 0:
    print("✓ No missing values in the dataset!")
else:
    print(missing[missing > 0])
```

```python
# Basic statistics
print("\n=== BASIC STATISTICS ===")
df.describe()
```

## Step 3: Understanding Our Target Variable (Churn)

```python
# Check churn distribution
print("\n=== CHURN DISTRIBUTION ===")
churn_counts = df['Churn'].value_counts()
print(churn_counts)
print(f"\nChurn Rate: {(df['Churn'] == 'Yes').sum() / len(df) * 100:.1f}%")
```

```python
# Visualize churn
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

# Bar chart
churn_counts.plot(kind='bar', ax=ax[0], color=['#2ecc71', '#e74c3c'])
ax[0].set_title('Customer Churn Distribution', fontsize=12, fontweight='bold')
ax[0].set_xlabel('Churn')
ax[0].set_ylabel('Count')
ax[0].set_xticklabels(['No', 'Yes'], rotation=0)

# Pie chart
churn_counts.plot(kind='pie', ax=ax[1], autopct='%1.1f%%', colors=['#2ecc71', '#e74c3c'])
ax[1].set_title('Churn Percentage', fontsize=12, fontweight='bold')
ax[1].set_ylabel('')

plt.tight_layout()
plt.show()

print("\n💡 Key Insight: About 28% of customers are churning. Our goal is to predict these 28% and intervene.")
```

## Step 4: Explore Relationships with Churn

### Question: Which factors are associated with churn?

```python
# Churn rate by Contract type (THIS IS A KEY PATTERN)
print("\n=== CHURN RATE BY CONTRACT TYPE ===")
contract_churn = df.groupby('Contract')['Churn'].apply(lambda x: (x == 'Yes').sum() / len(x) * 100)
print(contract_churn.sort_values(ascending=False))

# Visualize
plt.figure(figsize=(10, 5))
contract_churn.sort_values(ascending=False).plot(kind='bar', color=['#e74c3c', '#3498db', '#2ecc71'])
plt.title('Churn Rate by Contract Type', fontsize=12, fontweight='bold')
plt.ylabel('Churn Rate (%)')
plt.xlabel('Contract Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("\n💡 CRITICAL INSIGHT: Month-to-month contracts have ~40% churn vs. 2-year contracts with ~3% churn!")
print("This is THE biggest predictor of churn.")
```

```python
# Churn rate by Internet Service type
print("\n=== CHURN RATE BY INTERNET SERVICE ===")
internet_churn = df.groupby('InternetService')['Churn'].apply(lambda x: (x == 'Yes').sum() / len(x) * 100)
print(internet_churn.sort_values(ascending=False))

plt.figure(figsize=(10, 5))
internet_churn.sort_values(ascending=False).plot(kind='bar', color=['#e74c3c', '#3498db', '#2ecc71'])
plt.title('Churn Rate by Internet Service Type', fontsize=12, fontweight='bold')
plt.ylabel('Churn Rate (%)')
plt.xlabel('Internet Service')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Churn rate by Tenure (how long they've been a customer)
print("\n=== CHURN RATE BY TENURE ===")
# Group tenure into buckets
df['TenureBucket'] = pd.cut(df['tenure'], bins=[0, 6, 12, 24, 72], 
                             labels=['0-6 months', '6-12 months', '1-2 years', '2+ years'])
tenure_churn = df.groupby('TenureBucket')['Churn'].apply(lambda x: (x == 'Yes').sum() / len(x) * 100)
print(tenure_churn)

plt.figure(figsize=(10, 5))
tenure_churn.plot(kind='bar', color=['#e74c3c', '#f39c12', '#3498db', '#2ecc71'])
plt.title('Churn Rate by Customer Tenure', fontsize=12, fontweight='bold')
plt.ylabel('Churn Rate (%)')
plt.xlabel('Tenure')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("\n💡 INSIGHT: New customers (0-6 months) churn much more than loyal customers.")
print("This suggests we should focus retention efforts on early-stage customers.")
```

```python
# Churn by Senior Citizen status
print("\n=== CHURN RATE BY SENIOR CITIZEN STATUS ===")
senior_churn = df.groupby('SeniorCitizen')['Churn'].apply(lambda x: (x == 'Yes').sum() / len(x) * 100)
print(senior_churn)

plt.figure(figsize=(10, 5))
senior_churn.plot(kind='bar', color=['#3498db', '#e74c3c'])
plt.title('Churn Rate by Senior Citizen Status', fontsize=12, fontweight='bold')
plt.ylabel('Churn Rate (%)')
plt.xlabel('Senior Citizen (0=No, 1=Yes)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
